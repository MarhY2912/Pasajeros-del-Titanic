
Preprocesar los datos limpiándolos, tratando valores faltantes y transformándolos según sea necesario.

```python
# Rellenamos valores nulos de la variable 'Age' con la mediana
dataset['Age'].fillna(dataset['Age'].median(), inplace=True)

# Rellenamos valores nulos de la variable 'Embarked' con la moda
dataset['Embarked'].fillna(dataset['Embarked'].mode()[0], inplace=True)

# Eliminar filas con valores nulos en la variable 'Survived'
dataset.dropna(subset=['Survived'], inplace=True)
```

Conversión de variables categóricas

```python
# Convertimos las variables categóricas 'Sex' y 'Embarked' en variables numéricas
dataset['Sex'] = dataset['Sex'].map({'male': 0, 'female': 1})
dataset = pd.get_dummies(dataset, columns=['Embarked'], drop_first=True)
```

Seleccionar las características más relevantes para entrenar el modelo utilizando selección de características.

```python
#Análisis de la importancia de las características
from sklearn.ensemble import RandomForestClassifier

# Definimos las características (X) y la variable objetivo (y)
X = dataset.drop(columns=['Survived', 'Name', 'Ticket', 'Cabin'])  # Eliminamos columnas irrelevantes
y = dataset['Survived']

# Creamos el modelo de RandomForest
model = RandomForestClassifier()
model.fit(X, y)

# Mostramos la importancia de las características
feature_importances = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=False)
feature_importances.plot(kind='bar', figsize=(10, 6))
plt.title('Importancia de las Características')
plt.show()
```

Dividir el dataset en Train y Test para evaluar correctamente el modelo.

```python
from sklearn.model_selection import train_test_split

# Dividimos el dataset en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Entrenar el modelo configurando los diferentes hiperparámetros.

```python
from sklearn.linear_model import LogisticRegression

# Creamos y entrenamos el modelo
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
```

Evaluar el desempeño del modelo en el conjunto de Test con métricas como precisión, recall, F1-score, etc

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Predicciones del modelo
y_pred = logreg.predict(X_test)

# Evaluamos el rendimiento del modelo
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Imprimimos los resultados
print(f"Precisión: {accuracy:.2f}")
print(f"Precisión (Precision): {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
```

Realizar las diferentes gráficas que permitan visualizar los resultados del modelo.

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Matriz de confusión
cm = confusion_matrix(y_test, y_pred)

# Graficamos la matriz de confusión
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No Superviviente', 'Superviviente'], yticklabels=['No Superviviente', 'Superviviente'])
plt.title('Matriz de Confusión')
plt.ylabel('Real')
plt.xlabel('Predicción')
plt.show()
```

Curva ROC y AUC

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Calculamos la probabilidad de las predicciones positivas
y_prob = logreg.predict_proba(X_test)[:, 1]

# Calculamos la curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# Graficamos la curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'AUC = {roc_auc_score(y_test, y_prob):.2f}')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.title('Curva ROC')
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.legend(loc='lower right')
plt.show()
```

Precisión, Recall y F1-Score

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Calculamos las métricas
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Creamos una lista con los nombres y los valores de las métricas
metrics = ['Precisión', 'Recall', 'F1-Score']
values = [precision, recall, f1]

# Graficamos las métricas
plt.figure(figsize=(8, 6))
sns.barplot(x=metrics, y=values, palette='viridis')
plt.title('Métricas del Modelo')
plt.ylabel('Valor')
plt.show()
```

Importancia de las Características

```python
# Extraemos la importancia de las características
feature_importances = pd.Series(logreg.coef_[0], index=X.columns).sort_values(ascending=False)

# Graficamos la importancia de las características
plt.figure(figsize=(10, 6))
feature_importances.plot(kind='bar', color='teal')
plt.title('Importancia de las Características')
plt.ylabel('Importancia')
plt.show()
```

M
Matriz de Confusión:
La matriz de confusión es una de las herramientas más útiles para evaluar un modelo de clasificación. Este resultado muestra:

Verdaderos Positivos (TP): Número de pasajeros que sobrevivieron y fueron clasificados correctamente como sobrevivientes.

Falsos Negativos (FN): Número de pasajeros que sobrevivieron, pero el modelo los clasificó incorrectamente como no sobrevivientes.

Verdaderos Negativos (TN): Número de pasajeros que no sobrevivieron y fueron clasificados correctamente como no sobrevivientes.

Falsos Positivos (FP): Número de pasajeros que no sobrevivieron, pero el modelo los clasificó incorrectamente como sobrevivientes.
Un modelo ideal tendría valores elevados de TP y TN, y valores bajos de FP y FN.

Curva ROC y AUC:
La Curva ROC muestra la capacidad del modelo para clasificar correctamente entre las clases (sobreviviente/no sobreviviente) en diferentes umbrales.

El AUC (Área Bajo la Curva) es una métrica importante que nos indica qué tan bien el modelo puede distinguir entre las clases. Un valor de AUC cercano a 1 indica un excelente rendimiento, mientras que un valor cercano a 0.5 indica un modelo aleatorio.

Si el modelo tiene un AUC alto (por ejemplo, superior a 0.8), esto indica que el modelo es efectivo para distinguir entre los pasajeros que sobrevivieron y los que no.

Precisión: Nos indica qué porcentaje de las predicciones positivas fueron correctas. Una alta precisión es importante cuando los falsos positivos tienen un alto costo.

Recall: Nos indica qué porcentaje de los casos verdaderos fueron correctamente identificados. En este contexto, un alto recall es relevante porque queremos asegurarnos de que la mayoría de los sobrevivientes sean identificados.

F1-Score: Es la media armónica entre la precisión y el recall. Es útil cuando se desea un balance entre ambas métricas y es útil cuando se tienen clases desbalanceadas (por ejemplo, muchos más no sobrevivientes que sobrevivientes).
Un alto F1-Score indica que el modelo tiene un buen balance entre precisión y recall, lo que es esencial para este tipo de clasificación.

Precisión: Mide la exactitud de las predicciones positivas del modelo. Si la precisión es alta, el modelo hace pocas predicciones incorrectas para la clase positiva.

Recall: Evalúa qué tan bien el modelo encuentra todos los casos de la clase positiva. Si el recall es bajo, significa que el modelo no está identificando correctamente todos los sobrevivientes.

F1-Score: Es un buen indicador para un modelo balanceado, ya que toma en cuenta tanto la precisión como el recall.

AUC-ROC: Mide la capacidad del modelo para distinguir entre las dos clases (superviviente/no sobreviviente). Un valor cercano a 1 indica que el modelo es muy bueno en esa distinción.
